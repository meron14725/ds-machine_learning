## 線形回帰

機械学習では真の関数fを観測する($\hat{f}$を求める)

$$
\hat{Y} = \hat{f}(X)
$$
$\hat{Y}$ :予測値

$\hat{f}$ :真の関数の推測値

線形回帰では、fが”線形”と仮定する

$$
f(X) = \theta_{0}+\theta_{1}X_{1}+\theta_{2}X_{2}+\dots+\theta_{n}X_{n}
$$

$\theta_{j}$ :j番目のパラメータ

$X_{j}$ :j番目の特徴量の値

⇨最小二乗法で最適な$\theta$を見つける

## 最小二乗法

残差の二乗和が最小になるように線を引く


$$
\sum_{i=1}^{m} e^2_{i} = \sum_{i=1}^{m} \left( y_i - (\theta_0 + \theta_1 x_i) \right)^2
$$

⇨再急降下法により解く

## 再急降下法(gradient descent)

・関数の最小値を探索するアルゴリズム

・損失を表す関数を損失関数という



・最適化問題は、関数が最小になる変数の値を探索する

・最急降下法は最適化問題の勾配法のアルゴリズムの一つ

・最も急になる方向に少しずつパラメータを動かして最適解を探る（=学習）




## 最急降下法をスクラッチで実装する



```python
sample_df = pd.read_csv('data/sample_data.csv')
plt.plot(sample_df['space'],sample_df['rent'],'o')
```

```python
#損失関数を定義
x = sample_df['space'].values
y = sample_df['rent'].values
def cost_func(theta_0, theta_1, x, y):

    return np.mean(np.square(y-(theta_0+theta_1*x)))
```

```python
#損失関数の可視化
theta_0 = np.linspace(-20, 20, 300)
theta_1 = np.linspace(-1, 1 , 300)
X, Y = np.meshgrid(theta_0, theta_1)
z = [cost_func(param[0],param[1],x,y) for param in list(product(theta_0,theta_1))]
Z = np.array(z).reshape(len(theta_0),len(theta_1))
ax = plt.axes(projection='3d')
ax.plot_surface(X.T,Y.T,Z, cmap='jet')
plt.gca().invert_xaxis()
ax.view_init(elev=5, azim=0)
```

## 最急降下法

```python
theta_0_init = -5
theta_1_init = -0.5

def next_theta0(theta_0, theta_1, x, y, alpha=0.0005):
    return theta_0 - alpha*np.mean(theta_0+theta_1*x-y)*2

def next_theta1(theta_0, theta_1, x, y, alpha=0.0005):
    return theta_1 - alpha*np.mean((theta_0+theta_1*x-y)*x)*2

epochs = 100000
alpha = 0.00005
theta0_hist = []
theta1_hist = []

theta0_hist.append(theta_0_init)
theta1_hist.append(theta_1_init)

for _ in range(epochs):
    updated_theta0 = next_theta0(theta0_hist[-1], theta1_hist[-1], x=x,y=y,alpha=0.00005)
    updated_theta1 = next_theta1(theta0_hist[-1], theta1_hist[-1], x=x,y=y,alpha=0.00005)
    theta0_hist.append(updated_theta0)
    theta1_hist.append(updated_theta1)
```

```python
print(theta0_hist[-1])
print(theta1_hist[-1])
```

## 結果を描画

```python
##損失関数の推移の計算
theta_0_hist = np.array(theta0_hist)
theta_1_hist = np.array(theta1_hist)
cost_hist = [cost_func(param[0],param[1],x,y) for param in list(zip(theta_0_hist,theta_1_hist))]
ax = plt.axes(projection='3d')
ax.plot(theta_0_hist, theta_1_hist, cost_hist, 'x-')
ax.plot_surface(X.T,Y.T,Z, cmap='jet', alpha=0.5)
plt.gca().invert_xaxis()
ax.view_init(elev=30, azim=10)
```

```python
plt.contour(Y.T,X.T,Z, 100, cmap='jet')
plt.plot(theta_1_hist,theta_0_hist, '-')
```

```python
sns.scatterplot(x='space',y='rent',data=sample_df)
x_value = np.linspace(15, 120, 300)
y_value = theta_0_hist[-1] + theta_1_hist[-1]*x_value
plt.plot(x_value, y_value,'-', color = 'red')
```

## 正規方程式(normal equation)

・最急降下法を使わずに解析的に最適なパラメータ$\theta$を求める方程式

・一発で確実に解を求めることが可能

・線形回帰のモデルでのみ可能




m: 特徴量の種類

n: 特徴量の数



## 正規方程式をスクラッチで実装する

```python
#X
x_1 = sample_df['space'].values
x_0 = np.ones(len(x_1))
X_norm = np.column_stack((x_0,x_1))

#X.T*X -> XT_X
XT_X = np.dot(X_norm.T,X_norm)
#逆行列
XT_X_inv = np.linalg.inv(XT_X)
#theta
theta_result = np.dot(np.dot(XT_X_inv,X_norm.T),y)
```

```python
theta_result
```

```python
plt.contour(Y.T,X.T,Z, 100, cmap='jet')
plt.plot(theta_1_hist,theta_0_hist, '-')
plt.plot(theta_result[1],theta_result[0], 'o')
```

## 正規方程式の注意点

・特徴量が多すぎると$(\mathbf{X}^{\mathbf{T}}\mathbf{X})^{-1}$の計算に時間がかかる

・特徴量同士に強い相関があると$(\mathbf{X}^{\mathbf{T}}\mathbf{X})^{-1}$を求めることができない

・特徴量数がデータ数を上回る場合、回帰のパラメータを特定できない

## MSE

・ sklearn.metrics.mean_squared_error

    ・　mean_squared_error(y_true, y_pred)のように正解値と予測値のリストを渡す

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
df = sns.load_dataset('tips')
y_col = 'tip'
X = df.drop(columns=[y_col])
#標準化のために数値カラムのリストを取得
numeric_cols = X.select_dtypes(include=np.number).columns.to_list()
X = pd.get_dummies(X, drop_first=True)
y = df[y_col]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
#標準化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#学習データで特徴スケーリング
scaler.fit(X_train[numeric_cols])
X_train_scaled = X_train.copy()
#数値カラムのみ標準化
X_train_scaled[numeric_cols] = scaler.transform(X_train[numeric_cols])
X_test_scaled = X_test.copy()
#学習データのパラメータでテストデータを標準化
X_test_scaled[numeric_cols] = scaler.transform(X_test[numeric_cols])
```

```python
# 線形回帰モデル学習
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
```

```python
# モデルの評価(MSE)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
mean_squared_error(y_train,model.predict(X_train_scaled))
```
